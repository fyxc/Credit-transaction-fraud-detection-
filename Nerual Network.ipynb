```python
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
from numpy.random import seed

# set seed
seed(1)

# load dataset
train = pd.read_csv('train_selected.csv')
test = pd.read_csv('test_selected.csv')
oot = pd.read_csv('oot_selected.csv')
print(train.shape)
print(test.shape)
print(oot.shape)
```

```python
train.head()
```

```python
# train data
train_x = train.iloc[:,10:]
train_y = train.iloc[:,9]
# test data
test_x = test.iloc[:,10:]
test_y = test.iloc[:,9]
# hold out data
oot_x = oot.iloc[:,10:]
oot_y = oot.iloc[:,9]

train_x.head()
```

```python
# create model
model = Sequential()
model.add(Dense(15, input_dim=30, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='nadam', metrics=['accuracy'])
# Fit the model
model.fit(train_x, train_y, 
          validation_data=(test_x,test_y),epochs=20, batch_size=10, verbose=2)
# calculate predictions
predictions_train = model.predict(train_x)
```

```python
# combine the result to the original dataset in order to calculate the FDR
df_train_pre = pd.DataFrame(predictions_train) 
train_fdr = pd.concat([train,df_train_pre],axis = 1)
train_fdr.columns.values[-1]='prediction_score'
train_fdr_desc = train_fdr.sort_values('prediction_score', ascending = False)
train_fdr_desc.shape
train_fdr_top1 = train_fdr_desc.head(838)
```

```python
#train_fdr = train_fdr_top10['pre_result'].sum()/train_fdr_top10['fraud'].sum()
#train_fdr = train_fdr_top10['fraud'].sum()/train['fraud'].sum()
train_fdr = train_fdr_top1['Fraud'].sum()/train.Fraud.sum()
train_fdr
```

```python
# Do the same thing for test dataset
predictions_test = model.predict(test_x)
df_test_pre = pd.DataFrame(predictions_test) 
test_fdr = pd.concat([test,df_test_pre],axis = 1)
test_fdr.columns.values[-1]='prediction_score'
test_fdr_desc = test_fdr.sort_values('prediction_score', ascending = False)
test_fdr_top1 = test_fdr_desc.head(838)
#test_fdr_top10['pre_result'] = round(test_fdr_top10.prediction_score)
#test_fdr = test_fdr_top10['pre_result'].sum()/test_fdr_top10['fraud'].sum()
#test_fdr = test_fdr_top10['fraud'].sum()/test['fraud'].sum()
test_fdr = test_fdr_top1['Fraud'].sum()/test['Fraud'].sum()
test_fdr
```

```python
# Do the same thing for oot
predictions_oot = model.predict(oot_x)
df_oot_pre = pd.DataFrame(predictions_oot) 
oot_fdr = pd.concat([oot,df_oot_pre],axis = 1)
oot_fdr.columns.values[-1]='prediction_score'
oot_fdr_desc = oot_fdr.sort_values('prediction_score', ascending = False)
oot_fdr_top1 = oot_fdr_desc.head(252)
#oot_fdr_top10['pre_result'] = round(oot_fdr_top10.prediction_score)
#oot_fdr = oot_fdr_top10['pre_result'].sum()/oot_fdr_top10['fraud'].sum()
#oot_fdr = oot_fdr_top10['fraud'].sum()/oot['fraud'].sum()
oot_fdr = oot_fdr_top1['Fraud'].sum()/oot['Fraud'].sum()
oot_fdr
```

```python
train_fdr_desc.to_csv("train_pre.csv")
test_fdr_desc.to_csv("test_pre.csv")
oot_fdr_desc.to_csv("oot_pre.csv")
```
